SERIES

```python
#Crear una serie
serie = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
serie
```

```python
#Borrar series o filas
serie2 = serie.drop(['a','e'])
serie2
```

```python
#Seleccionar filas o series
serie[['b','d']]
```

```python
#Asignación
serie['f']=200
serie
```

```python
#Operadores
serie4 = serie+serie2
serie4
```

```python
#Multiplicación
serie4 = serie4.mul(serie, serie2)
serie4
```

```python
# División
serie4 = serie4.div(serie, serie2)
serie4
```

```python
serie4 = serie4.sub(serie, serie2)
serie4
```

DATAFRAMES

```python
datos = {'Alumno':['Mar','Lucia','Juan','Eva'],'Nota':[20,15,18,10]}
df = pd.DataFrame(datos,columns=['Alumno','Nota'])
df
```

```python
#Borrar los datos
df2 = df.drop([0,2])
df2
```

```python
#Eliminar la columna alumno
df3 = df.drop(['Alumno'], axis=1)
df3
```

```python
#Cambiar la norta Eva
df.loc[3,'Nota']=17
df
```
